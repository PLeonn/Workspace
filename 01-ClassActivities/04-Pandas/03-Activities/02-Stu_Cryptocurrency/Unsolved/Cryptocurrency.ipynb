```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merge_df = pd.merge(bitcoin_df,dash_df,on="Date",how="inner")
merge_df.head()
```

```python
# Rename columns so that they are differentiated
merge_df = merge_df.rename(columns={"Open_x": "Open Bitcoin",
                                                "High_x": "High Bitcoin",
                                                "Low_x": "Low Bitcoin",
                                                "Close_x": "Close Bitcoin",                                                "High_x": "Hair Color",
                                                "Volume_x": "Volume Bitcoin",
                                                "Market Cap_x": "Mkt Cap Bitcoin",
                                                "Open_y": "Open Dash",
                                                "High_y": "High Dash",
                                                "Low_y": "Low Dash",
                                                "Close_y": "Close Dash",                                                "High_x": "Hair Color",
                                                "Volume_y": "Volume Dash",
                                                "Market Cap_y": "Mkt Cap Dash"
                                                })
merge_df.head()
```

```python
# alternatively you can set your suffixes when the merge occurs
merge_df = pd.merge(bitcoin_df,dash_df,on="Date",how="inner",suffixes=("_Bitcoin","_Dash"))
merge_df.head()
```

```python
# Collecting best open for Bitcoin and Dash
openB = merge_df["Open_Bitcoin"].max()
openD = merge_df["Open_Dash"].max()

# Collecting best close for Bitcoin and Dash
closeB = merge_df["Close_Bitcoin"].max()
closeD = merge_df["Close_Dash"].max()


# Collecting the total volume for Bitcoin and Dash
volB = merge_df["Volume_Bitcoin"].max()
volD = merge_df["Volume_Dash"].max()

print(f"{openB}, {openD}, {closeB}, {closeD}, {volB}, {volD}")
```

```python
# Creating a summary DataFrame using above values
summaryDF = pd.DataFrame({"Best BC Open":[openB],
                          "Best Dash Open":openD,
                          "Best BC Close":closeB,
                          "Best Dash Close":closeD,
                          "Best BC Volume":volB,
                          "Best Dash Volume":volD})
summaryDF.head()
```
